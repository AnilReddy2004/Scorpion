```python
from sklearn.datasets import make_circles

n_samples=1000
X,y=make_circles(n_samples,
                 noise=0.04,
                 random_state=42)
```

```python
import pandas as pd
df=pd.DataFrame({"X1":X[:,0],"X2":X[:,1],"y":y})
df.head()
```

```python
from matplotlib import pyplot as plt
plt.scatter(df.X1,df.X2,c=y)
plt.show()
```

```python
import tensorflow as tf
tf.__version__
```

```python
tf.random.set_seed(42)

model=tf.keras.Sequential([
    tf.keras.layers.Dense(10,activation="relu"), # Added more units and activation
    tf.keras.layers.Dense(1,activation="sigmoid") # Changed to 1 output unit and sigmoid activation
])

model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
              optimizer=tf.keras.optimizers.SGD(learning_rate=0.02),
              metrics=["accuracy"])

model.fit(X,y,epochs=100)
```

```python
model.evaluate(X,y)
```

```python
#Lets build a another model to improve from the above model

tf.random.set_seed(42)

m2=tf.keras.Sequential([
    tf.keras.layers.Dense(10,activation="relu"),
    tf.keras.layers.Dense(10,activation="relu"),
    tf.keras.layers.Dense(1,activation="sigmoid")
])

m2.compile(loss=tf.keras.losses.BinaryCrossentropy(),
           optimizer=tf.keras.optimizers.SGD(learning_rate=0.02),
           metrics=["accuracy"])

m2.fit(X,y,epochs=150)
```

```python
m2.evaluate(X,y)
```

```python
m2.predict(X)
```
